```python
numeric_features = [feature for feature in df.columns if df[feature].dtype == int]
categorical_features = [feature for feature in df.columns if df[feature].dtype != int]

print(f'We have {len(numeric_features)} numerical features : {numeric_features}')
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average_score'] = (df['total_score'] / 3).round(2)
df.head()
```

```python
reading_score_full = df[df['reading_score'] == 100].shape[0]
math_score_full = df[df['math_score'] == 100].shape[0]
writing_score_full = df[df['writing_score'] == 100].shape[0]

print(f'Number of students with full score in reading : {reading_score_full}')
print(f'Number of students with full score in math : {math_score_full}')   
print(f'Number of students with full score in writing : {writing_score_full}')
```

```python
reading_less_20 = df[df['reading_score'] < 20].shape[0]
math_less_20 = df[df['math_score'] < 20].shape[0]
writing_less_20 = df[df['writing_score'] < 20].shape[0]

print(f'Number of students with less than 20 score in reading : {reading_less_20}')
print(f'Number of students with less than 20 score in math : {math_less_20}')   
print(f'Number of students with less than 20 score in writing : {writing_less_20}')
```
